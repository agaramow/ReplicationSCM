# How Synthetic control method works? 

Synthetic control methods is aimed to cunstruct a convex combination of multiple control units that best approximates the relevant characteristics of the treated unit during the pretreatment period. That is the method contstructs synthetic unit which best approximates the treated unit before an intervention. Then the post intervention outcomes for the synthetic control unit are used to estimate the outcomes of the treated unit in the absence of the intervention. 

Abadie and Gardeazabal (2003), Abadie, Diamond, and Hainmueller(2010), and Abadie, Diamond, and Hainmueller (2013) introduced a statistical software [Synth](https://web.stanford.edu/~jhain/synthpage.html) that implements synthetic control methods for causal inference in comparative case studies with aggregate data as described in )

_Synth_ package is avilable for MATLAB, Stata, and R. In the current notebook we use the Synth package intended for R with the use of [rpy2](https://rpy2.readthedocs.io/en/version_2.8.x/overview.html#background) package, which allows to call R functions from Python. 

# How is Synthetic control group constructed? 

The model for constructing Sythetic control group is descreibed in Abadie, Diamond, and Hainmueller(2010) and Abadie,Diamond, and Hainmueller (2011). 

Suppose $j=1, ..., J+1$ units are observed for time periods $t=1,...,T$. Without loss of generality assume that only the first unit is exposed to the intervention. That is $j=1$ is treatment unit and $j=2,...,J+1$ are potential control units that can contribute to the synthetic control. The intervention occurs at time period $T_0 +1$, so that $1,2,...T_0$ are pre intervention periods and $T_0+1,T_0+2,...,T$ are post intervention periods. 

Two potential outcomes: $Y_{it}^0$ is the outcome that would be observed for unit $i$ at time $t$ in the absence of the intervention, and $Y_{it}^1$ is the outcome when $i$ is exposed to the intervention in periods $T_0+1 \text{ to } T$. 

**Assumptions**: 
1. The intervention has no effect on the outcome before implementation period, that is for $t\in{1,...,T}$ and $i\in{1,...N}$, we have that $Y_{it}^N=Y_{it}^I$.   
### conditional independence, ignorability
2. Outcomes of the untreated units are not affected by the intervention implemented in the treated unit. 
### Sutva

There might be cases that intervention affects the outcome before intervention date (anticipation effects), in such cases we can redefine $T_0$ anticipation date. In the anticiapation effects part of the effect of financial derugulation to the top income share in the UK, we change the treatment date from 1986 (actual intervention date) to 1983 (anticipation date, when the law about financial deregualtion was approved) and check whether there were anticipation effects.

<font color="red"> Let $Y_{it}^1$ be the outcome for the treated unit, UK, in the treated state and  $Y_{it}^0$ the outcome of UK in the untreated state. It is clear that in the pre-treatment period $Y_{it}^1=Y_{it}^0$. After treatment is introduced $Y_{it}^1!=Y_{it}^0$ </font>

$\alpha_{it}=Y_{it}^1-Y_{it}^0$ be the *ATT* for unit $i$ at time $t$.  The observed outcome for unit $i$ at time $t$ is $Y_{it}=Y_{it}^0+\alpha _{it}D_{it}$. Since only the first region is exposed to the intervention after period $T_0$, we have 
\begin{align*}
D_{it} = \begin{cases}
1 & \quad\text{if}\quad i=1 \text{ and } t>T_0 \\
0 & \quad\text{if}\quad \text{otherwise}
\end{cases}
\end{align*} 

As mentioned above our goal is to estimate the *ATT* in the post-intervention period: $\alpha_{1T_0+1},..., \alpha_{1T}$. $\alpha_{1t}=Y_{1t}^I-Y_{1t}^N=Y_{1t}-Y_{1t}^N$ for $T_0+1,T_0+2,...,T$ periods. Since $Y_{it}^N$ is unobserved on the treated unit in the post-intervention period. Abadie, Diamond, and Hainmueller(2010) consider $Y_{it}^N$ as given by a factor model:
\begin{equation}
Y_{it}^N=\delta_t+\bf{\theta_t Z_i}+\lambda_t U_i+\epsilon_{it}
\end{equation} 

where $\delta_t$ is unknown common factor loadings across units, $\bf{Z_i}$ is a $(r \times 1)$ vector of observed covariates (not affected by the intervention) <font color="red"> explain how Z is chosen, difference between dif and dif </font>, $\bf{\theta_t}$ is a $(r \times 1)$ of vector of unknown parameters, $\lambda_t$ is a vector of unobserved common factors, $U_i$ is a vector of unknown factor loadings,  $\epsilon_{it}$ is unobserved transitory shocks. $Z_i$ can be constructed by choosing combinations of covariates and years, for instance weighted average over some years. The Synthetic control method does not limit the functional form, since Synthetic control method optimizes given data. The functional form above is provided to show the difference between dif and dif and synthetic control method.

Ideally, we would like to constuct synthetic control group that resembles the treated unit in all relevant pre-intervention characteristics. Formalizing this idea, Abadie, Diamond, and Hainmueller(2010) takes a $(T_0\times 1)$ vector $K=(k_1, k_2, ... , k_{T_0})'$ that gives some linear combinations of pre-intervention outcomes: $\bar{Y}_i^K=\sum_{s=1}^{T_0} {k_s,Y_{is}}$. Linear combinations of pre-intervention outcomes van be used to observe unobserved common factors whose effect vary over time. One can choose up to $T_0$ linear combinations, $M$, to fit the comparison units based on the outcome variables in the pre-intervention period. 

To construct the sythetic control group, Abadie, Diamond, and Hainmueller(2010) define a vector of weights ${\bf W}=(w_2,...w_{J+1})$ such that $w_j\geq 0$ and $w_2+...+w_{J+1}=1$. The authors propose to choose the optimal weight $\bf W^*$ such that synthetic control unit best approximates the unit exposed to the intervention with respet to the outcome predictors $Z_i$ and $M$ linear combinations of pre-intervention outcomes $\bar{Y}_i^{K_1},...,\bar{Y}_i^{K_M}$ such that $\sum_{j=2}^{J+1}w_j^* {\bar{Y}_j^{K_1}}=\bar{Y}_1^{K_1},...,\sum_{j=2}^{J+1}w_j^* {\bar{Y}_j^{K_M}}=\bar{Y}_1^{K_M} \text{ and } \sum_{j=2}^{J+1}w_j^* Z_j=Z_1$ hold, then an estimator for $\alpha_{1t}$ is 

\begin{align} \hat{\alpha}_{1t}=Y_{1t}-\sum_{j=2}^{J+1} w_j^* Y_{jt} \text{ for } t\in\{T_0+1,...,T\} \end{align}

In applications it might be the case that there is no $W^*$ such that $ \sum_{j=2}^{J+1} w_j^* \bar{Y}_j^{K_1}=\bar{Y}_1^{K_1},...,\sum_{j=2}^{J+1}w_j^* {\bar{Y}_j^{K_M}}=\bar{Y}_1^{K_M}$  and $ \sum_{j=2}^{J+1} w_j^* Z_j=Z_1$ hold exactly, in such cases weights will be choosen to hold the equations approximately. The optimal weights should minimize the distance between the synthetic control unit and the treated unit. To define the distance authors combine characteristics of the exposed unit in $k\times 1$ ($k=r+M$) matrix $X_1=(Z_1',\bar{Y}_1^{K_1},...,\bar{Y}_1^{K_M})'$ and for the conrol units in $k\times J$ matrix $X_0$. Then $W^*$ minimizes 
\begin{align} ||X_1-X_0 W||_V=\sqrt{(X_1-X_0 W)'V(X_1-X_0 W)} \tag{1} 
\end{align} 

Where $V$ is defined as some $k\times k$ symmetric and positive simidefinite matrix. $V$ allows different weights to the variables in $X_0$ and $X_1$ depending on their predictive power on outcome. An optimal choice of $V$ assigns weights that minimize the mean square error of the synthetic control estimator, that is $(Y_1-Y_0W^*)'(Y_1-Y_0W^*)$. Synth() function that we will use below uses data driven procedure to choose $V^*$, which is choosen among all positive definite and diagonal matrices such that MSPE of the outcome variable is minimized over some set of pre-intervention periods. Let $Q_1$ be a $(T_P\times 1)$ vector of th outcome variable for the treated unit and $Q_0$ be the $(T_P\times J)$ analogous matrix for the control units, where $T_P (1\leq T_P \leq T_0)$ is the number of preintervention perods over which MSPE is minimized. Note that the rows of $Q_0$ and $Q_1$ form subsets of $Y_0$ and $Y_1$ respectively. Actually we could choose $Y_0$ and $Y_1$ when $T_P = T_0$, however mostly it is suficient to choose $T_P < T_0$ to achieve a low MSPE over the entire pre-treatment period. Then $V^*$ is chosen to minimize 

\begin{align}
arg \min_{V}((Q_1-Q_0 W^*(V))'(Q_1-Q_0 W^*(V)) \tag{2}
\end{align}

Synth() function solves a nested optimization problem that minimizes equation (2) given by equation (1)



# to delete

$\sum_{j=2}^{J+1} w_j Y_{jt}=\delta_t+{\bf\theta_t}\sum_{j=2}^{J+1} w_j {\bf Z_j} +\lambda_t \sum_{j=2}^{J+1} w_j \mu_j+\sum_{j=2}^{J+1} w_j \epsilon_{jt}$. 

Suppose that there are $(w_2^*,...w_{J+1}^*)$ such that  \begin{align*} 
\sum_{j=2}^{J+1} w_j Y_{j1}=Y_{1 1}, \sum_{j=2}^{J+1} w_j Y_{j2}=Y_{1 2}, ... ,\sum_{j=2}^{J+1} w_j Y_{jT_0}=Y_{1 T_0} \text{ and } 
\sum_{j=2}^{J+1} w_j {\bf Z}_j={\bf Z_1}\end{align*} 

Then an estimator for $\alpha_{1t}$ is $\hat{\alpha_{1t}}=Y_{1t}-\sum_{j=2}^{J+1} w_j^* Y_{jt}$ for $t\in\{T_0+1,...,T\}$
# How Syth package works
The synth() function allows for exibility in the choice of V . Sometimes the researcher has a good subjective assessment of the predictive power of the variables in X1 and X0. In this
case the user could supply his own weights via the custom.V option. These weighs populate the diagonal of the V matrix (with the non-diagonal elements equal to zero) and synth() simply minimizes equation (1) conditional on the user supplied V.

#### Using dataprep function of R we prepare the dataset to use in synth function of "Synth" package

Using dataprep() frunction of the Synth package, we reorginize the panel dataset into an appropriate format that is suitable for the main estimator synth(). synth() requires as an input four data matrices $X_1, X_0, Z_1, Z_0$ that are needed to construct synthetic control group. $X_1$ is the vector of predictors for the treted unit, $X_0$ matrix of predictors for the control units, $Z_1$ vector of ourcome variable for treated unit and $Z_0$ is matrix of outcome for control units. 

Synth() command searches for the vecor of weights $W^*$ that identifies the synthetic control for treated unit. For any $V$ synth() finds a $W_*(V)$ minimizing equation of distance 

```python
# Here we converted dataprep() function in R to Python
# inputs are 1) dataframe (should be R dataframe), 2) predictor_variables, 3) time_predictors_pr - time period applied to predictors 4)dependent_variable 5) treated_unit 6)control_units
# 7) time period over which to optimize 8) time_pl - time period for plotting outcome

def dataprep(dataframe, predictor_variables, time_predictors_pr, dependent_variable, treated_unit, control_units, time_optimize, time_plo):
    dataprep_out=r_synth.dataprep(foo = dataframe,
    predictors = StrVector((predictor_variables)),
    predictors_op = "mean",
    time_predictors_prior = IntVector(time_predictors_pr),
    dependent = dependent_variable, 
    unit_variable = "ccode",
    unit_names_variable="Country",
    time_variable = "year",
    treatment_identifier = treated_unit,
    controls_identifier = IntVector(control_units),
    time_optimize_ssr = time_optimize,
    time_plot = time_plo)
    return dataprep_out

def synth(dateprep_out):
    return r_synth.synth(data_prep_obj = dateprep_out, method = "BFGS")

def synth_tables(dateprep_out, synth_out):
    return r_synth.synth_tab(dataprep_res = dataprep_out, synth_res = synth_out)

def plot_path(dataprep_out, synth_out, tr_date, period):
    plt.plot(period, dataprep_out.rx2("Y0plot").dot(synth_out.rx2("solution.w")), label='SCG' )
    plt.plot(period, dataprep_out.rx2("Y1plot"), label='treated unit')
    plt.axvline(x=tr_date, color="grey")
    plt.xlabel('Year')
    plt.ylabel('Income share, %')
    plt.legend()
    return plt.show()
    
def weights(dataprep_out, synth_out):
    df1= pandas2ri.ri2py(dataprep_out.rx2("names.and.numbers"))[1:]
    df1=df1.reset_index(drop=True)
    df2=pd.DataFrame(np.around(pandas2ri.ri2py((synth_out.rx2("solution.w"))), decimals=3).tolist())
    df1['weights']=df2
    df1.rename(columns={'unit.names':'Country', 'unit.numbers':'unit numbers'}, inplace=True)
    return df1

def gaps(dataprep_out, synth_out):
    gaps = dataprep_out.rx2("Y1plot").ro-(dataprep_out.rx2("Y0plot").dot(synth_out.rx2("solution.w")))
    return gaps

def gaps_percentage(dataprep_out1, synth_out1):
    x1=pandas2ri.ri2py(dataprep_out1.rx2("Y1plot").ro-(dataprep_out1.rx2("Y0plot").dot(synth_out1.rx2("solution.w"))))
    x2=pandas2ri.ri2py(dataprep_out1.rx2("Y1plot"))
    gaps_percentage= np.divide(x1,x2)
    return gaps_percentage

def plot_gaps(dataprep_out, synth_out, tr_date, period):
    gaps = dataprep_out.rx2("Y1plot").ro-(dataprep_out.rx2("Y0plot").dot(synth_out.rx2("solution.w")))
    plt.plot(period, gaps)
    plt.ylabel('Gaps in nominal values')
    plt.xlabel('Year')
    plt.axhline(y=0, color='black', linestyle='--')
    plt.axvline(x=tr_date, color="grey")
    return plt.show()
    
def plot_gaps_percentage(dataprep_out, synth_out, tr_date, period):
    x1=pandas2ri.ri2py(dataprep_out.rx2("Y1plot").ro-(dataprep_out.rx2("Y0plot").dot(synth_out.rx2("solution.w"))))
    x2=pandas2ri.ri2py(dataprep_out.rx2("Y1plot"))
    gaps_percentage= np.divide(x1,x2)
    plt.plot(period, gaps_percentage)
    plt.ylabel('Gaps, %')
    plt.xlabel('Year')
    plt.axhline(y=0, color='black', linestyle='--')
    plt.axvline(x=tr_date, color="grey")
    return plt.show()
```
